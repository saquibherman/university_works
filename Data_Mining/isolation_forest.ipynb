```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.ensemble import IsolationForest
data = pd.read_csv('joined_df.csv')
data.head(10)
```

```python

model=IsolationForest(n_estimators=100,max_samples='auto',contamination=float(0.2))

model.fit(data)

print(model.get_params())
```

```python
data['scores'] = model.decision_function(data)

data['anomaly_score'] = model.predict(data)

data[data['anomaly_score']==-1].head()
```

```python
data[data['anomaly_score']==-1]
```

```python
accuracy = 100*list(data['anomaly_score']).count(-1)/(len(data[data['anomaly_score']==-1]))
print("Accuracy of the model:", accuracy)
```
